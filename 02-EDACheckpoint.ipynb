# Research Question

How does emotional and cognitive intensity during autobiographical memory recall, measured by self-rated stress, mental drain, sentiment, and valence change, influence narrative detail and recall effort, measured by word count, memory duration, and writing time, and does this relationship differ between individuals ages 18–25 vs. 68-70?





## Background and Prior Work

Emotion has a significant impact on cognitive processes in humans, including attention, learning, and memory. The valence (positive or negative) and intensity (high or low) are differently related to autobiographical memories. 1 The emotional valence associated with autobiographical memories is important because positive memories tend to be recalled more frequently than negative memories. Previous research has found that autobiographical memories used for social functions tended to be positive, while memories recalled to direct future behaviors and decisions were the least positive ones.

In a study that experimented with autobiographical memories that varied in valence and intensity, it was found that the emotional intensity affects the properties of autobiographical memories more than valence. 2 The study concluded that highly intense events tend to be remembered longer and with greater vividness.

In a study on emotional intensity in episodic autobiographical memory and counterfactual thinking, it was found that emotional intensity during the remembered event was significantly higher for negative than positive events. 3 However, when remembering emotional events, there were no significant differences in emotional intensity between negative and positive memories. Additionally, the study evaluates episodic counterfactual thought, which is the act of imagining alternative ways that a past memory might have occurred. Memories can either be imagined to be better or worse than the events occurred, and accompanied by intense emotions. It was found that there was no difference in emotional intensity when remembering and emotional intensity participants imagined they would have during counterfactual events.

Another study has also found that a perspective change from a field (first-person) to an observer (third-person) reduces the emotional intensity of recalled events. 4 Given the study above, a field-to-observer perspective change is likely to reduce the vividness of a memory. Furthermore, the study indicates that the autobiographical memory system has plasticity, and a perspective change alters the autobiographical records of recalled events.

^ Tabea Wolf, Justina Pociunaite, Sophie Hoehne, Daniel Zimprich, The valence and the functions of autobiographical memories: Does intensity matter?, Consciousness and Cognition, Volume 91, 2021, 103119, ISSN 1053-8100, https://doi.org/10.1016/j.concog.2021.103119.
^ Talarico, J.M., LaBar, K.S. & Rubin, D.C. Emotional intensity predicts autobiographical memory experience. Memory & Cognition 32, 1118–1132 (2004). https://doi.org/10.3758/BF03196886
^ Matthew L. Stanley, Natasha Parikh, Gregory W. Stewart, Felipe De Brigard, Emotional intensity in episodic autobiographical memory and counterfactual thinking, Consciousness and Cognition, Volume 48, 2017, Pages 283-291, ISSN 1053-8100, https://doi.org/10.1016/j.concog.2016.12.013.
^ Sekiguchi, T., & Nonaka, S. (2014). The long-term effect of perspective change on the emotional intensity of autobiographical memories. Cognition and Emotion, 28(2), 375–383. https://doi.org/10.1080/02699931.2013.825233

# Hypothesis


Using an emotional intensity scale of 1-5, with no 0 values, we hypothesize that narratives reported as high in emotional intensity (≥4) will be associated with greater vividness, more sensory detail, and more frequent use of a first-person perspective in the text, compared to narratives with lower and medium intensity ratings (0–3). This is based on the premise that experiences deemed more important or stressful to retell enhance memory encoding and recording of events, preserving detailed sensory and perspective information in autobiographical narratives.
