## Abstract

Emotion plays an important role in cognitive processes such as attention, learning, and memory, with both valence (positive vs. negative) and intensity shaping autobiographical recall. Prior research suggests that emotionally intense events are remembered with greater vividness and duration, while positive memories are more frequently recalled. However, the relationship between emotional intensity and measurable aspects of recall, such as story duration, remains unclear. 

In this project, we analyze two datasets to examine this relationship. Using Hippocorpus data, we compare self-reported emotional intensity (stressful and draining ratings) with word count, writing time, and first-person perspective. To do this, we group observations into low/medium ratings (1-3) and high (ratings 4-5) intensity levels based on the self-reported ratings, and perform group comparisons and a nonparametric statistical test (Mann-Whitney U Test). To further examine how emotion relates to recall duration, we use the Remembering Past Emotions dataset to compare memory duration and emotional valence across happy and sad memories using a Wilcoxon signed-ranked test. 

Results from Hippocorpus show that high-intensity memories have significantly higher word counts and writing times than low/medium intensity memories (p < 0.001), indicating high-intensity memories involve greater narrative detail and recall effort, while first-person perspective remains unchanged. We found that in the Remembering Past Emotions Dataset emotional valence differs significantly between happy and sad memories for both past and present ratings (p < 0.001), whereas memory duration does not differ significantly between cue types (p = 0.821). 

Overall, these findings suggest that emotional intensity influence the richness and effort of memory recall, while emotional valence shapes how memories are experienced without affecting their duration.

## Background and Prior Work

Emotion has a significant impact on cognitive processes in humans, including attention, learning, and memory. The valence (positive or negative) and intensity (high or low) are differently related to autobiographical memories. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The emotional valence associated with autobiographical memories is important because positive memories tend to be recalled more frequently than negative memories. Previous research has found that autobiographical memories used for social functions tended to be positive, while memories recalled to direct future behaviors and decisions were the least positive ones.

In a study that experimented with autobiographical memories that varied in valence and intensity, it was found that the emotional intensity affects the properties of autobiographical memories more than valence. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) The study concluded that highly intense events tend to be remembered longer and with greater vividness.

In a study on emotional intensity in episodic autobiographical memory and counterfactual thinking, it was found that emotional intensity during the remembered event was significantly higher for negative than positive events. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) However, when remembering emotional events, there were no significant differences in emotional intensity between negative and positive memories. Additionally, the study evaluates episodic counterfactual thought, which is the act of imagining alternative ways that a memory might have occurred. Memories can either be imagined to be better or worse than the events that occurred, and be accompanied by intense emotions. It was found that there was no difference in emotional intensity when remembering and the emotional intensity participants imagined they would have during counterfactual events.

Another study has also found that a perspective change from a field (first-person) to an observer (third-person) reduces the emotional intensity of recalled events. <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Given the study above, a field-to-observer perspective change is likely to reduce the vividness of a memory. Furthermore, the study indicates that the autobiographical memory system has plasticity, and a perspective change alters the autobiographical records of recalled events.

# Analysis

## Hippocorpus

This section examines whether memories rated as higher in emotional intensity differ from lower and medium intensity memories in narrative detail and recall effort. To do this, we group observations into low/medium (ratings 1–3) and high (ratings 4–5) intensity based on self-reported emotional ratings.

#### Comparing intensity groups to Word Count, First-person rate, and writing time

```python
# Create combined intensity variable
hippo['intensity_combined'] = np.where(
    (hippo['draining'] >= 4) | (hippo['stressful'] >= 4),
    "High",
    "Low/Medium"
)

# Check distribution
print(hippo['intensity_combined'].value_counts())

# Create first-person rate
hippo['first_person_rate'] = hippo['first_person_count'] / hippo['word_count']

# Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Word count
sns.boxplot(data=hippo, x='intensity_combined', y='word_count', ax=axes[0])
axes[0].set_title('Word Count by Intensity')

# First-person rate
sns.boxplot(data=hippo, x='intensity_combined', y='first_person_rate', ax=axes[1])
axes[1].set_title('First-Person Rate by Intensity')

# Writing time
sns.boxplot(data=hippo, x='intensity_combined', y='workTimeInSeconds', ax=axes[2])
axes[2].set_title('Writing Time by Intensity')

plt.tight_layout()
plt.show()
```

High-intensity memories show slightly higher median word counts and writing times compared to low/medium intensity memories, suggesting a small increase in narrative detail and recall effort. First-person rate appears nearly identical across groups, indicating that narrative perspective remains consistent regardless of emotional intensity. Overall, while there are modest upward shifts for high-intensity memories, the substantial overlap in distributions suggests that differences are relatively small.

To determine whether the observed differences between low/medium and high intensity memories are statistically meaningful, we compare the two groups on word count, first-person rate, and writing time using Mann–Whitney U tests. This test is appropriate because the outcome distributions are skewed and contain many outliers.

```python
from scipy.stats import mannwhitneyu

# Split groups
low_med = hippo[hippo['intensity_group'] == 'Low/Medium']
high = hippo[hippo['intensity_group'] == 'High']

# Variables to test
variables = ['word_count', 'first_person_rate', 'workTimeInSeconds']

results = []

for var in variables:
    group1 = low_med[var].dropna()
    group2 = high[var].dropna()

    n1 = len(group1)
    n2 = len(group2)
    
    stat, p = mannwhitneyu(group1, group2, alternative='two-sided')

    # z-score
    mu_U = (n1 * n2) / 2
    sigma_U = np.sqrt((n1 * n2 * (n1 + n2 + 1)) / 12)
    Z = (stat - mu_U) / sigma_U
    
    # effect size r
    r = Z / np.sqrt(n1 + n2)
    
    results.append({
        'variable': var,
        'low_medium_median': group1.median(),
        'high_median': group2.median(),
        'u_statistic': stat,
        'p_value': p,
        'effect_size': abs(r)
    })

results_df = pd.DataFrame(results)
results_df
```

The Mann–Whitney U tests show that high-intensity memories take slightly longer to write than low or medium intensity memories (p = .003), but the effect size is very small (r.036). This means that while the difference is statistically significant, it is not very meaningful in practice. There are no significant differences in word count (p = .149) or first-person rate (p = .725), suggesting that memory intensity does not meaningfully affect how long people write or the perspective they use. Overall, the distributions still overlap heavily, so the effects are weak and likely not practically important.

### Emotional Valence

This section examines whether emotional memory characteristics differ between happy and sad cue conditions within the 17 participants in the data. To do this, we reshape the data into a long format, where each row represents the log_duration values of one participant, and perform a Wilcoxon signed-rank test.

```python
wide_df = emotions_df.pivot_table(index='participant',
                                  columns='cue_word',
                                  values='log_duration',
                                  aggfunc='mean')
```

## Discussion and Conclusion

Our project explored how emotional intensity during autobiographical memory recall influences narrative detail and recall effort. Overall, the results show that emotional intensity has a measurable but small effect on how people recall and describe memories. In the Hippocorpus dataset, higher-intensity memories were associated with slightly higher word counts and longer writing times, suggesting increased recall effort. However, the boxplots and distributions showed a large overlap between high and low/medium intensity groups, and the statistical effect sizes were very small. This indicates that while emotional intensity does have an effect, it does not meaningfully change how detailed people’s narratives are. Additionally, first-person perspective remained consistent across intensity levels, showing no significant relationship with emotional intensity.

These findings both support and extend prior research. Previous studies suggested that emotional intensity increases the vividness and strength of autobiographical memories. Our results partially support this, as more intense memories required more effort to recall, shown by longer writing times. However, they challenge the idea that these memories are significantly more detailed. One explanation is that past research often measured subjective vividness, while our study used observable measures like word count. This suggests that emotional intensity may make memories feel more vivid without actually increasing the amount of detail people express.

The second dataset adds another layer by examining emotional valence and memory duration. The results show that happy and sad memories differ significantly in emotional valence, both in the past and present, but do not differ in how long they are recalled. The visualizations support this, showing similar distributions of memory duration across cue types. At the same time, past emotions were more extreme, while present emotions were closer to neutral. This suggests that emotional experiences change over time, supporting the idea that memory is reconstructive rather than fixed.

Taken together, these results suggest that emotional intensity affects the effort of recall more than the actual content or structure of memory narratives. This helps explain why emotional memories often feel more vivid or important, even if they are not more detailed when measured objectively.

There are also some contradictions in the findings. For example, while high-intensity memories took longer to write, they did not use more first-person language, which might be expected if they were more immersive. Additionally, negative memories are often thought to last longer, but our analysis showed no difference in duration between happy and sad memories. These inconsistencies may be due to differences between internal experience and how people express memories in writing.

There are several limitations to this study. Both datasets rely on self-reported data, which can introduce bias. Measures like word count and writing time are only indirect indicators of memory detail. The data is also cross-sectional, meaning it captures memories at one point in time rather than tracking changes over time. Additionally, demographic imbalances in the datasets may limit how generalizable the results are.

Future work could improve this by using more advanced text analysis methods to better capture narrative detail, and by collecting longitudinal data to study how memories and emotions change over time. Combining self-reported data with more objective measures of emotion could also strengthen the findings.

In conclusion, emotional intensity plays a role in memory recall, but its effects are subtle. While it increases recall effort slightly, it does not strongly impact narrative detail or perspective. This suggests that emotional memories may feel more vivid without actually being more detailed.
