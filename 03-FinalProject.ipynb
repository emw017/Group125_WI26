## Research Question

How does emotional and cognitive intensity during autobiographical memory recall, measured by self-rated stress, mental drain, sentiment, and valence change, influence narrative detail and recall effort, measured by word count, memory duration, and writing time?

## Background and Prior Work

Emotion has a significant impact on cognitive processes in humans, including attention, learning, and memory. The valence (positive or negative) and intensity (high or low) are differently related to autobiographical memories. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The emotional valence associated with autobiographical memories is important because positive memories tend to be recalled more frequently than negative memories. Previous research has found that autobiographical memories used for social functions tended to be positive, while memories recalled to direct future behaviors and decisions were the least positive ones.

In a study that experimented with autobiographical memories that varied in valence and intensity, it was found that the emotional intensity affects the properties of autobiographical memories more than valence. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) The study concluded that highly intense events tend to be remembered longer and with greater vividness.

In a study on emotional intensity in episodic autobiographical memory and counterfactual thinking, it was found that emotional intensity during the remembered event was significantly higher for negative than positive events. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) However, when remembering emotional events, there were no significant differences in emotional intensity between negative and positive memories. Additionally, the study evaluates episodic counterfactual thought, which is the act of imagining alternative ways that a memory might have occurred. Memories can either be imagined to be better or worse than the events that occurred, and be accompanied by intense emotions. It was found that there was no difference in emotional intensity when remembering and the emotional intensity participants imagined they would have during counterfactual events.

Another study has also found that a perspective change from a field (first-person) to an observer (third-person) reduces the emotional intensity of recalled events. <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Given the study above, a field-to-observer perspective change is likely to reduce the vividness of a memory. Furthermore, the study indicates that the autobiographical memory system has plasticity, and a perspective change alters the autobiographical records of recalled events.

## Exploring the Use of First Person Perspective 

```python
# Since there is no column for this, we'll loop through each participant's story for the count!
participant_stories = hippo['story'].str.split().tolist()

first_person = ['i', 'my', 'me', 'mine', 'we', 'us', 'ours', 'our'] 
first_counts = []

for words in participant_stories:
    counter = 0
    for word in words:
        if word.lower() in first_person:
            counter += 1
    first_counts.append(counter)
hippo['first_person_count'] = first_counts
```

## Hippocorpus - Analysis

This section examines whether memories rated as higher in emotional intensity differ from lower and medium intensity memories in narrative detail and recall effort. To do this, we group observations into low/medium (ratings 1–3) and high (ratings 4–5) intensity based on self-reported emotional ratings.

```python
# Create binary intensity group based on *draining ratings
def intensity_group(x):
    if x >= 4:
        return "High"
    else:
        return "Low/Medium"

hippo['intensity_group'] = hippo['draining'].apply(intensity_group)

hippo['intensity_group'].value_counts()
```

#### Comparing intensity groups to Word Count, First-person rate, and writing time

```python
# First-person rate (normalize by length)
hippo['first_person_rate'] = hippo['first_person_count'] / hippo['word_count']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Word count
sns.boxplot(data=hippo, x='intensity_group', y='word_count', ax=axes[0])
axes[0].set_title('Word Count by Intensity')

# First-person rate
sns.boxplot(data=hippo, x='intensity_group', y='first_person_rate', ax=axes[1])
axes[1].set_title('First-Person Rate by Intensity')

# Writing time
sns.boxplot(data=hippo, x='intensity_group', y='workTimeInSeconds', ax=axes[2])
axes[2].set_title('Writing Time by Intensity')

plt.tight_layout()
plt.show()
```

High-intensity memories show slightly higher median word counts and writing times compared to low/medium intensity memories, suggesting a small increase in narrative detail and recall effort. First-person rate appears nearly identical across groups, indicating that narrative perspective remains consistent regardless of emotional intensity. Overall, while there are modest upward shifts for high-intensity memories, the substantial overlap in distributions suggests that differences are relatively small.

To determine whether the observed differences between low/medium and high intensity memories are statistically meaningful, we compare the two groups on word count, first-person rate, and writing time using Mann–Whitney U tests. This test is appropriate because the outcome distributions are skewed and contain many outliers.

```python
from scipy.stats import mannwhitneyu

# Split groups
low_med = hippo[hippo['intensity_group'] == 'Low/Medium']
high = hippo[hippo['intensity_group'] == 'High']

# Variables to test
variables = ['word_count', 'first_person_rate', 'workTimeInSeconds']

results = []

for var in variables:
    group1 = low_med[var].dropna()
    group2 = high[var].dropna()
    
    stat, p = mannwhitneyu(group1, group2, alternative='two-sided')
    
    results.append({
        'variable': var,
        'low_medium_median': group1.median(),
        'high_median': group2.median(),
        'u_statistic': stat,
        'p_value': p
    })

results_df = pd.DataFrame(results)
results_df
```

The Mann–Whitney U tests show that high-intensity memories have significantly higher word counts and writing times than low/medium intensity memories (p < 0.001), suggesting modest increases in narrative detail and recall effort. There is no significant difference in first-person rate (p = 0.41), indicating narrative perspective remains consistent. Overall, the differences are statistically significant but small, consistent with the overlap in distributions.

### Emotional Valence - Analysis

This section examines whether emotional memory characteristics differ between happy and sad cue conditions within the 17 participants in the data. To do this, we reshape the data into a long format, where each row represents the log_duration values of one participant, and perform a Wilcoxon signed-rank test.

```python
wide_df = emotions_df.pivot_table(index='participant',
                                  columns='cue_word',
                                  values='log_duration',
                                  aggfunc='mean')
```

```python
wide_df = wide_df.dropna()
```

```python
from scipy.stats import wilcoxon

stat, p = wilcoxon(wide_df['Happy'], wide_df['Sad'])

print("statistic:", stat)
print("p-value:", p)
```

These results indicate that there is no significant difference (p=0.821). We fail to reject the null hypothesis that emotional memory duration differs between happy and sad cues. 

```python
variables = ['log_duration', 'past_valence', 'present_valence']

for var in variables:
    wide = emotions_df.pivot_table(index='participant',
                                   columns='cue_word',
                                   values=var,
                                   aggfunc='mean').dropna()
    
    stat, p = wilcoxon(wide['Happy'], wide['Sad'])
    print(f"{var}: p = {p:.6f}")
```

These results show that there were significant differences in both past valence and present valence (p < 0.001). These findings suggest that cue type strongly influences the emotional valence of memories, but it does not significantly impact the duration of those memories.
