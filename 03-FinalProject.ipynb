## Research Question

How does emotional and cognitive intensity during autobiographical memory recall, measured by self-rated stress, mental drain, sentiment, and valence change, influence narrative detail and recall effort, measured by word count, memory duration, and writing time?

## Background and Prior Work

Emotion has a significant impact on cognitive processes in humans, including attention, learning, and memory. The valence (positive or negative) and intensity (high or low) are differently related to autobiographical memories. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The emotional valence associated with autobiographical memories is important because positive memories tend to be recalled more frequently than negative memories. Previous research has found that autobiographical memories used for social functions tended to be positive, while memories recalled to direct future behaviors and decisions were the least positive ones.

In a study that experimented with autobiographical memories that varied in valence and intensity, it was found that the emotional intensity affects the properties of autobiographical memories more than valence. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) The study concluded that highly intense events tend to be remembered longer and with greater vividness.

In a study on emotional intensity in episodic autobiographical memory and counterfactual thinking, it was found that emotional intensity during the remembered event was significantly higher for negative than positive events. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) However, when remembering emotional events, there were no significant differences in emotional intensity between negative and positive memories. Additionally, the study evaluates episodic counterfactual thought, which is the act of imagining alternative ways that a memory might have occurred. Memories can either be imagined to be better or worse than the events that occurred, and be accompanied by intense emotions. It was found that there was no difference in emotional intensity when remembering and the emotional intensity participants imagined they would have during counterfactual events.

Another study has also found that a perspective change from a field (first-person) to an observer (third-person) reduces the emotional intensity of recalled events. <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Given the study above, a field-to-observer perspective change is likely to reduce the vividness of a memory. Furthermore, the study indicates that the autobiographical memory system has plasticity, and a perspective change alters the autobiographical records of recalled events.

### Exploratory Data Analysis on The Hippocorpus Data

Now that we wrangle our data, let's explore some distributions throughout it and see if we can discover any underlying relationships!


```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
hippo = pd.read_csv("data/02-processed/hippoCorpus_processed.csv")
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, col in zip(axes, ['stressful', 'draining', 'distracted']):
    sns.countplot(data=hippo, x=col, ax=ax)
    ax.set_title(f'Distribution of {col}')
plt.tight_layout()
```

The distributions of responses for stress, drain, and distraction variables follow a right-skewed distribution. This implies that most participants reported a less 'extreme' storytelling experience, whereas outliers with more 'extreme' responses when recalling their story pull the mean of each distribution to the right.

### Word Count
Since the word count of the stories is important to us but is not already in the dataset, we will need to add it ourselves.

```python
hippo['word_count'] = hippo['story'].str.split().str.len()

plt.figure(figsize=(8,4))
sns.histplot(hippo['word_count'], bins=50)
plt.title('Distribution of Story Word Count')
plt.xlabel('word_count')
plt.show()
```

The distribution of word count appears roughly normal with a slight right tail. Most participants wrote stories between about 150 and 300 words, while a smaller group wrote much longer responses, which creates the extended tail on the right side of the distribution.

## Word_count vs The Three Intensity Ratings (Stressful, Draining, Distracted)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

sns.boxplot(data=hippo, x='stressful', y='word_count', ax=axes[0])
axes[0].set_title('Word Count by Stress Level')
axes[0].set_xlabel('stressful')
axes[0].set_ylabel('word_count')

sns.boxplot(data=hippo, x='draining', y='word_count', ax=axes[1])
axes[1].set_title('Word Count by Draining Level')
axes[1].set_xlabel('draining')
axes[1].set_ylabel('')

sns.boxplot(data=hippo, x='distracted', y='word_count', ax=axes[2])
axes[2].set_title('Word Count by Distracted Level')
axes[2].set_xlabel('distracted')
axes[2].set_ylabel('')

plt.tight_layout()
plt.show()
```

# Exploratory Data Analysis on Remembering Past Emotions Dataset

For this data set, we will get a closer look at what particular emotions can tell us about the recall of autobiographical events!

```python
emotions_df = pd.read_csv("data/02-processed/emotions_dataset.csv")
emotions_df['log_duration'] = np.log1p(emotions_df['emotional_memory_duration'])
emotions_df.head(2)
```

### Emotional Memory Duration Distribution

```python
plt.figure(figsize=(8,2))
sns.histplot(data= emotions_df, x= 'emotional_memory_duration')
```

The histogram shows a strongly right-skewed distribution of emotional memory duration, with most memories concentrated between 20 and 100 units, with a few extending beyond 200, and an extremely low number of outliers extending beyond 300. This suggests that while most emotional experiences are relatively short-lived, a minority persist for much longer periods, potentially reflecting highly significant or emotionally intense events. The range and spread of the histogram indicate a wide variability in emotional memory duration, indicating emotional experiences differ widely in how long they are remembered or emotionally processed. The presence of a long right tail in the distribution definitely tells us that certain emotional experiences are encoded and retained for longer durations, which may correspond to high-intensity memories or stories that are expected to exhibit greater vividness, sensory detail, and first-person perspective. #edit this add the influence of different emotions as well

### Duration vs Emotion

```python
# Since the distribution of the duration was skewed, we will log as an attempt to normalize
emotions_df['log_duration'] = np.log1p(emotions_df['emotional_memory_duration'])
# We will store this in another column within our dataset 
plt.figure(figsize=(8,4))
plot_emotions = sns.boxplot(data= emotions_df, y= 'log_duration', x= 'cue_word')

average_duration = emotions_df.groupby('cue_word')['log_duration'].mean().reset_index(name='mean')
# The log transformation lessens the skewness and becomes more symmetrical, making the mean as a form of central tendency more appropriate
average_duration

#This shows the average log duration of emotional memories for each cue word
```

The bar chart compares the average log-transformed emotional memory duration for memories prompted by happy vs sad cue words. Memories associated with sad cues show a very slightly higher average duration, ~3.89, compared to happy cues, ~3.85. This suggests that negative emotional memories may persist slightly longer than positive ones. However, the difference between the two groups is small, indicating that the overall duration of emotional memories is relatively similar across cue types. 

### Valence Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(10,2))

sns.histplot(emotions_df['past_valence'], bins=20, ax=axes[0])
sns.histplot(emotions_df['present_valence'], bins=20, ax=axes[1])
axes[0].set_title('Past Valence')
axes[1].set_title('Present Valence')
```

We notice a stark difference between the distribution of the two histograms measuring past and present valence. The past valence histogram shows a bimodal distribution, indicating that participants likely felt stronger positive/negative emotions when recalling how they felt during past memories. On the other hand, the present valance histogram shows the mode at 0, indicating that most participants felt indifferent when recalling past memories.

### Past / Present Valence vs Duration

```python
bins = [1.5, 2.5, 3.5, 4.5, 5.5, 6.5]
emotions_df["log_duration_groups"] = pd.cut(emotions_df["log_duration"], bins=bins)


df_long = emotions_df.melt(
    id_vars="log_duration_groups",
    value_vars=["past_valence", "present_valence"],
    var_name="valence_type",
    value_name="valence")

ax = sns.boxplot(
    data=df_long,
    y="log_duration_groups",
    x="valence",
    hue="valence_type")

ax.set_title('Valence by Log Duration Times')

sns.move_legend(
    ax,
    "center left",
    bbox_to_anchor=(1, 0.5),
    title='Valence Type')

print('Past Valence Median:', emotions_df['past_valence'].median())
print('Present Valence Median:', emotions_df['present_valence'].median())
```

In this boxplot, we notice that the values in the present_valence group seem to consistenly more positive in comparison to the values in the past_valence group, which seem to be more evenly distributed across the positive and negative valence values. The past_valence group shows more variability. These findings seem to be relatively consistent across duration groups.
